# **What is Artificial Intelligence , Machine Learning & Deep Learning ?**

image.png


AI, ML, and DL are abbreviations commonly used in the field of artificial intelligence. Here's a brief explanation of each term:

1. AI (Artificial Intelligence): AI refers to the simulation of human intelligence in machines that are programmed to think and learn like humans. It involves creating intelligent systems capable of performing tasks that typically require human intelligence, such as problem-solving, decision-making, speech recognition, and natural language processing.

2. ML (Machine Learning): Machine Learning is a subfield of AI that focuses on the development of algorithms and statistical models that enable computers to learn from and make predictions or decisions based on data without being explicitly programmed. ML algorithms learn patterns and insights from large datasets and use them to make accurate predictions or take actions.

3. DL (Deep Learning): Deep Learning is a subset of machine learning that utilizes artificial neural networks inspired by the structure and function of the human brain. Deep learning algorithms are designed to automatically learn hierarchical representations of data by processing multiple layers of interconnected neurons. It is particularly effective for handling large and complex datasets, such as images, speech, and text, and has achieved remarkable success in areas like computer vision, natural language processing, and speech recognition.

In summary, AI is the broader field encompassing the creation of intelligent systems, ML is a subfield of AI that focuses on algorithms and statistical models for learning from data, and DL is a subset of ML that employs deep neural networks to process and learn from complex data.

# **Train Test Split**

image.png

```python
y = salary_data['Salary']
x = salary_data[['Experience Years']]
```

```python
#step 4 split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
#step5 select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#step6 train model
model.fit(x_train,y_train)
model.intercept_
model.coef_
```

```python
#step7 predict model
y_pred = model.predict(x_test)
```

```python
#step8 : accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
